```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import numpy as np
```

```python
data = pd.read_csv("Student_Performance.csv")
data
```

`We want to know how a Sleep Hours affects Performance Index?`

so we say that :


`Sleep Hours` is `independent` and is the `feature` 
while the `Performance Index` are `dependent` and are the `target`

```python

# we clean the unessisary column

data.drop(["Hours Studied",'Previous Scores','Extracurricular Activities','Sample Question Papers Practiced'],axis="columns", inplace=True)
data
```

```python
# we need to rename columns
data.rename(columns={"Sleep Hours": "sleep_hours", "Performance Index": "performance"},inplace=True)
```

```python
x = pd.DataFrame(data, columns = ["sleep_hours"])
y = pd.DataFrame(data, columns = ['performance'])

data.describe()
```

```python
plt.scatter(x, y, alpha = 0.009, c="orange")

plt.xlabel("Sleep Hours")
plt.ylabel("Performance")

plt.xlim(3.0, 10.0)
plt.ylim(10.0, 100.0)
```

```python
reg = LinearRegression()
```

```python
reg.fit(x, y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
plt.scatter(x, y, alpha = 0.009, c="orange")

plt.xlabel("Sleep Hours")
plt.ylabel("Performance")


plt.plot(x, reg.predict(x), color='red')

plt.xlim(3.0, 10.0)
plt.ylim(10.0, 100.0)
```

### Pridicting Values using Linear regression


```python
sleep_hours = 5.0 


sleep_hours_array = np.array([[sleep_hours]])

predicted_performance = reg.predict(sleep_hours_array)

predicted_performance[0][0]
```
